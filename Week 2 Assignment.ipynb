## Part 1- Data Coding and Merging

```python
#Import Libraries

import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import NHL Team Data and Display
NHL_Team=pd.read_csv("Assignment Data/Week 2/NHL_team.csv")
display(NHL_Team)
```

```python
#Drop columns
NHL_Team.drop(['Unnamed: 0', 'abbr', 'tname', 'lname', 'sname'], axis=1, inplace=True)
display(NHL_Team)
```

```python
#Rename team name column
NHL_Team.rename(columns={'name':'team_name'}, inplace=True)
display(NHL_Team)
```

```python
#Import NHL Competition Data and Display Head
NHL_Competition=pd.read_csv("Assignment Data/Week 2/NHL_competition.csv")
NHL_Competition.head()
```

```python
#Drop columns
NHL_Competition.drop(['Unnamed: 0', 'tz', 'start', 'end'], axis=1, inplace=True)
display(NHL_Competition)
```

```python
#Rename columns
NHL_Competition.rename(columns={'name':'competition_name'}, inplace=True)
display(NHL_Competition)
```

```python
#Import NHL Game Data and Display Head
NHL_Game=pd.read_csv("Assignment Data/Week 2/NHL_game.csv")
NHL_Game.head()
```

```python
#Drop columns
NHL_Game.drop(['X', 'period', 'status'], axis=1, inplace=True)
display(NHL_Game)
```

```python
#Merge NHL_Game & NHL_Team dataframes
NHL_Game=pd.merge(NHL_Team, NHL_Game, on = ['tid'])
NHL_Game.head()
```

```python
#Merge NHL_Game & NHL_Competition dataframes
NHL_Game=pd.merge(NHL_Competition, NHL_Game, on = ['comp_id'])
NHL_Game.head()
```

```python
#Create variable for goal differential
NHL_Game['hgd']=NHL_Game['hscore']-NHL_Game['ascore']
```

```python
#Remove missing data
NHL_Game=NHL_Game[pd.notnull(NHL_Game['hgd'])]
NHL_Game.shape
```

```python
#Convert date to datetime
NHL_Game['date']=pd.to_datetime(NHL_Game['date'])
NHL_Game['date'].head()
```

```python
#Sort games by date, display first 15 results
NHL_Game.sort_values(by=['date'], ascending=[False]).head(15)
```

```python
#Separate into home & away databases
#Label GF/GA columns
NHL_Home=NHL_Game[NHL_Game.home_away != 'away']
NHL_Home.rename(columns={'ascore':'goals_against'}, inplace=True)
NHL_Home.rename(columns={'hscore':'goals_for'}, inplace=True)
NHL_Home
```

```python
#Create a win column
NHL_Home['win']=np.where(NHL_Home['hgd']>0, 'W', 'L')
NHL_Home['games_played']=1
NHL_Home
```

```python
#Separate into home & away databases
#Label GF/GA columns
NHL_Away=NHL_Game[NHL_Game.home_away != 'home']
NHL_Away.rename(columns={'ascore':'goals_for'}, inplace=True)
NHL_Away.rename(columns={'hscore':'goals_against'}, inplace=True)
NHL_Away
```

```python
#Create a win column
NHL_Away['win']=np.where(NHL_Away['hgd']<1, 'W', 'L')
NHL_Away['games_played']=1
NHL_Away
```

```python
#Append the home & away databases
NHL_Game=pd.concat([NHL_Home, NHL_Away], ignore_index=True)
NHL_Game
```

```python
dummy=pd.get_dummies(NHL_Game, columns=['win'])
dummy.columns
```

```python
NHL_Game=pd.concat([NHL_Game, dummy['win_W']], axis=1)
NHL_Game.head()
```

```python
#Create a team level database to show GF/GA broken down by regular season and playoffs in each season
NHL_Team_Stats=NHL_Game.groupby(['team_name', 'competition_name', 'comp_id', 'type'])['team_name','goals_for','goals_against','win_W'].sum()
display(NHL_Team_Stats)
```

```python
NHL_Team_Stats=NHL_Team_Stats.reset_index()
display(NHL_Team_Stats)
```

```python
#Create a database to show the number of games played by each team in the regular season and playoffs each season
NHL_Game_Count=NHL_Game.groupby(['team_name','competition_name', 'comp_id', 'type'])['games_played'].sum()
NHL_Game_Count
```

```python
#Merge the two databases
NHL_Team_Stats=pd.merge(NHL_Team_Stats, NHL_Game_Count, on=['team_name','competition_name','type'])
display(NHL_Team_Stats)
```

```python
NHL_Team_Stats=pd.merge(NHL_Team, NHL_Team_Stats, on = ['team_name'])
display(NHL_Team_Stats)
```

```python
#Import NHL PPPK Data and Display Head
NHL_PPPK=pd.read_csv("Assignment Data/Week 2/pp.pk.ppgf.csv")
NHL_PPPK.head()
```

```python
NHL_Team_Stats=pd.merge(NHL_PPPK, NHL_Team_Stats, on = ['tricode','competition_name'])
display(NHL_Team_Stats)
```

```python
NHL_Team_Stats['win_pct']=NHL_Team_Stats['win_W']/NHL_Team_Stats['games_played']
NHL_Team_Stats['avg_gf']=NHL_Team_Stats['goals_for']/NHL_Team_Stats['games_played']
NHL_Team_Stats['avg_ga']=NHL_Team_Stats['goals_against']/NHL_Team_Stats['games_played']
display(NHL_Team_Stats)
```

```python
NHL_Team_R_Stats=NHL_Team_Stats[NHL_Team_Stats.type !=3]
display(NHL_Team_R_Stats)
```
